# 1. Building an ensemble of neurons

**a) Tuning curves.** Plot the population tuning curves. Plot the representation accuracy plot ($x - \hat{x}$). Compute and report the RMSE.

```python
def RMSE(x, xhat):
    error = x - xhat
    return np.sqrt(np.power(error,2).mean())
```

```python
# ✍ 
N = 100
tau_RC = 0.02
tau_Ref = 0.0002

model = nengo.Network()
with model: 
    ensemble = nengo.Ensemble( N,
                               dimensions = 1,
                               max_rates = nengo.dists.Uniform(100,200),
                               neuron_type = nengo.LIF(tau_rc = tau_RC, tau_ref = tau_Ref)
                              )
    connection = nengo.Connection(ensemble, ensemble)

with nengo.Simulator(model) as simulator:
    x, activities = tuning_curves(ensemble, simulator)
    decoder = simulator.data[connection].weights.T

xhat = np.dot(activities,decoder)

error = RMSE(x, xhat)
print("RMSE:\n")
print(error)

plt.figure(figsize=(5,5))
plt.plot(x,activities, label = "tuning curves")
plt.ylabel("Firing Rate (Hz)")
plt.xlabel("Represented value x")

plt.figure(figsize=(5,5))
plt.plot(x,x-xhat, label = "x - $\hat{x}$")
plt.legend()
```

# 2. Connecting neurons

**a) Computing the identity function.** Show the input value and the decoded values from the two  ensembles in three separate plots. Run the simulation for $0.5\,\mathrm{s}$.

```python
# 
tau_RC = 0.02
tau_Ref = 0.0002

model = nengo.Network()
with model: 
    ensemble1 = nengo.Ensemble( 100,
                               dimensions = 1,
                               max_rates = nengo.dists.Uniform(100,200),
                               neuron_type = nengo.LIF(tau_rc = tau_RC, tau_ref = tau_Ref)
                              )
    ensemble2 = nengo.Ensemble( 50,
                               dimensions = 1,
                               max_rates = nengo.dists.Uniform(100,200),
                               neuron_type = nengo.LIF(tau_rc = tau_RC, tau_ref = tau_Ref)
                              )
    stimulus = nengo.Node(lambda t: 0 if t<= 0.1 else (1 if t <= 0.4 else 0))  
    
    connection1 = nengo.Connection(stimulus, ensemble1)
    connection2 = nengo.Connection(ensemble1, ensemble2,  synapse = 0.01)
    
    ps = nengo.Probe(stimulus)
    p1 = nengo.Probe(ensemble1,  synapse = 0.01)
    p2 = nengo.Probe(ensemble2,  synapse = 0.01)
    
with nengo.Simulator(model) as simulator:
    simulator.run(0.5)
    plt.figure()
    plt.title("")
    plt.plot(simulator.trange(), simulator.data[p1])
    plt.xlabel('t')
    plt.ylabel('estimate')

    plt.figure()
    plt.title("")
    plt.plot(simulator.trange(), simulator.data[p2])
    plt.xlabel('t')
    plt.ylabel('estimate')

    plt.figure()
    plt.title("")
    plt.plot(simulator.trange(), simulator.data[ps], c='k')
    plt.xlabel('t')
    plt.ylabel('stimulus')
```

**b) Integrator using spiking neurons.**  Show the input, the ideal integral, and the value represented by the ensemble when the input is a value of $0.9$ from $t=0.04$ to $t=1.0$ (and $0$ for other times). Run the simulation for $1.5\,\mathrm{s}$.

```python
# ✍ 
def integrate(N, d, r, r1, r2, neuronT, tau1, tau2, inputF, T):  

    model = nengo.Network()
    with model: 
        ensemble = nengo.Ensemble( N,
                                   dimensions = d,
                                   radius = r,
                                   max_rates = nengo.dists.Uniform(r1,r2),
                                   neuron_type = neuronT
                                  )
        stimulus = nengo.Node(inputF)
        connection1 = nengo.Connection(stimulus, 
                                       ensemble, 
                                       transform = tau2,  
                                       synapse = tau1)
        connection2 = nengo.Connection(ensemble, 
                                       ensemble,  
                                       synapse = tau2)

        ps = nengo.Probe(stimulus)
        p1 = nengo.Probe(ensemble,  synapse = 0.01)
       
    with nengo.Simulator(model) as simulator:
        simulator.run(T)
        idealI = np.cumsum(simulator.data[ps])*T/len(simulator.data[ps])
        plt.figure()
        plt.title("")
        plt.plot(simulator.trange(), simulator.data[p1], label = "integrator")
        plt.plot(simulator.trange(), simulator.data[ps], label = "stimulus")
        plt.plot(simulator.trange(),idealI, label = "ideal integrator")
        plt.xlabel('t')
        plt.ylabel('estimate')
        plt.legend()
```

```python
N = 200
d = 1
r = 1
r1 = 100
r2 = 200
tau_RC = 0.02
tau_Ref = 0.002
neuronT = nengo.LIF(tau_rc = tau_RC, tau_ref = tau_Ref)
tau1 = 0.005
tau2 = 0.05
inputF = lambda t: 0 if t<= 0.04 else (0.9 if t <= 1.0 else 0)
T = 1.5

integrate(N, d, r, r1, r2, neuronT, tau1, tau2, inputF, T)
```
